# Bayes Rule Example Notebook

This notebook calculates posterior probabilities using Bayes' theorem.

```python
# Define the prior probability distribution
prior = {
    'A': 0.3,
    'B': 0.2,
    'C': 0.5
}

# Define the likelihoods
likelihood = {
    'A': 0.8,
    'B': 0.5,
    'C': 0.2
}

# Calculate the evidence or normalization constant
evidence = sum(prior[k] * likelihood[k] for k in prior.keys())

# Calculate the posterior probabilities
posterior = {k: prior[k] * likelihood[k] / evidence for k in prior.keys()}

# Print the results
print("Prior: ", prior)
print("Likelihood: ", likelihood)
print("Evidence: ", evidence)
print("Posterior: ", posterior)
```
