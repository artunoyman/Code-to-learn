# Ridge Regression Implementation and Analysis

This notebook implements Ridge Regression from scratch and demonstrates its application on a synthetic dataset. The implementation includes:
- Data generation with controlled noise
- Ridge Regression implementation with regularization
- Lambda selection using Leave-One-Out Cross-Validation (LOOCV)
- Visualization of results and model behavior


```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Data Generation
We'll generate synthetic data with two features:
- X1: Uniformly distributed between 1 and 10
- X2: A noisy multiple of X1
- Y: Target variable with higher relevance to X1

```python
# Seed for reproducibility
np.random.seed(100)

# Number of samples
n_samples = 50

# Generate Feature 1 (X1) uniformly between 1 and 10
X1 = np.random.uniform(1, 10, n_samples)

# Generate Feature 2 (X2) as a noisy multiple of X1
noise_X2 = np.random.normal(0, 2.0, n_samples)  # Increased noise
X2 = 3 * X1 + noise_X2

# Generate Target Variable (Y) with higher relevance to X1
noise_Y = np.random.normal(0, 3.0, n_samples)  # Increased noise
Y = 5 + 4 * X1 + noise_Y  # Increased coefficient for X1 to make it more relevant

# Combine features into a single matrix and add intercept term
X = np.column_stack((np.ones(n_samples), X1, X2))

# Display first 5 data points
print("First 5 modified data points:")
for i in range(5):
    print(f"Data Point {i+1}: X1={X1[i]:.2f}, X2={X2[i]:.2f}, Y={Y[i]:.2f}")
```

## 2. Ridge Regression Implementation
Implementation of the Ridge Regression algorithm with regularization parameter Î».

```python
def ridge_regression(X, Y, lambd):
    """
    Computes Ridge Regression coefficients.

    Parameters:
    X (numpy.ndarray): Feature matrix with intercept term.
    Y (numpy.ndarray): Target vector.
    lambd (float): Regularization parameter.

    Returns:
    numpy.ndarray: Coefficients vector.
    """
    XtX = np.dot(X.T, X)
    
    # Regularization matrix (do not regularize the intercept)
    n_features = X.shape[1]
    regularization = lambd * np.eye(n_features)
    regularization[0, 0] = 0  # No regularization for intercept
    
    # Compute (X^T X + lambda * I)
    XtX_reg = XtX + regularization
    
    # Compute X^T Y
    XtY = np.dot(X.T, Y)
    
    # Invert (X^T X + lambda * I)
    try:
        XtX_reg_inv = np.linalg.inv(XtX_reg)
    except np.linalg.LinAlgError:
        print(f"Matrix is singular for lambda={lambd}. Using pseudo-inverse.")
        XtX_reg_inv = np.linalg.pinv(XtX_reg)
    
    # Compute coefficients
    beta = np.dot(XtX_reg_inv, XtY)
    return beta
```

## 3. Mean Squared Error Function
Function to compute the Mean Squared Error between predicted and actual values.

```python
def mse(Y_true, Y_pred):
    """
    Computes Mean Squared Error.

    Parameters:
    Y_true (numpy.ndarray): Actual target values.
    Y_pred (numpy.ndarray): Predicted target values.

    Returns:
    float: Mean Squared Error.
    """
    return np.mean((Y_true - Y_pred) ** 2)
```

## 4. Lambda Selection via LOOCV
Implementation of Leave-One-Out Cross-Validation for selecting the optimal regularization parameter.

```python
def select_lambda_loocv(X, Y, lambdas):
    """
    Selects the best lambda using Leave-One-Out Cross-Validation.

    Parameters:
    X (numpy.ndarray): Feature matrix with intercept term.
    Y (numpy.ndarray): Target vector.
    lambdas (list): List of lambda values to evaluate.

    Returns:
    float: Best lambda value.
    list: Corresponding MSEs for each lambda.
    """
    best_lambda = None
    lowest_mse = float('inf')
    mse_list = []
    
    for lambd in lambdas:
        mse_total = 0
        for i in range(len(Y)):
            # Create training data by excluding the i-th data point
            X_train = np.delete(X, i, axis=0)
            Y_train = np.delete(Y, i)
            X_val = X[i].reshape(1, -1)
            Y_val = Y[i]
            
            # Train Ridge Regression
            beta = ridge_regression(X_train, Y_train, lambd)
            
            # Predict the validation point
            Y_pred = np.dot(X_val, beta).item()  # Extract scalar value
            
            # Compute MSE
            mse_total += (Y_val - Y_pred) ** 2
        
        avg_mse = mse_total / len(Y)
        mse_list.append(avg_mse)
        print(f"Lambda: {lambd}, LOOCV MSE: {avg_mse:.4f}")
        
        if avg_mse < lowest_mse:
            lowest_mse = avg_mse
            best_lambda = lambd
            
    return best_lambda, mse_list
```

## 5. Finding the Best Lambda
Let's find the optimal regularization parameter using LOOCV.

```python
# Define a range of lambda values (logarithmic scale for better coverage)
lambdas = [0.01, 0.1, 1, 10, 100, 1000]

# Select the best lambda using LOOCV
best_lambda, mse_list = select_lambda_loocv(X, Y, lambdas)

print(f"Best lambda selected via LOOCV: {best_lambda} with LOOCV MSE: {min(mse_list):.4f}")
```

## 6. Final Model with Best Lambda
Computing the final coefficients using the best lambda value.

```python
# Compute final coefficients with best lambda
final_beta = ridge_regression(X, Y, best_lambda)
print("Final Ridge Regression Coefficients:")
print(f"Intercept: {final_beta[0]:.4f}")
print(f"Beta1 (X1): {final_beta[1]:.4f}")
print(f"Beta2 (X2): {final_beta[2]:.4f}")

# Predict Y using the final model
Y_pred = np.dot(X, final_beta)

# Compute Training MSE
training_mse = mse(Y, Y_pred)
print(f"Training MSE on Modified Dataset: {training_mse:.4f}")
```

## 7. Visualization
Let's create various plots to visualize our results:
1. Features vs. Target
2. Predicted vs. Actual Values
3. Coefficient Paths
4. Regression Line

```python
# A. Scatter Plots of Features vs. Target
plt.figure(figsize=(14, 6))

# Plot Feature 1 vs. Y
plt.subplot(1, 2, 1)
plt.scatter(X1, Y, color='blue', label='Data Points')
plt.xlabel('Feature X1')
plt.ylabel('Target Y')
plt.title('Feature X1 vs. Target Y')
plt.legend()
plt.grid(True)

# Plot Feature 2 vs. Y
plt.subplot(1, 2, 2)
plt.scatter(X2, Y, color='green', label='Data Points')
plt.xlabel('Feature X2')
plt.ylabel('Target Y')
plt.title('Feature X2 vs. Target Y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# B. Predicted vs. Actual Y
plt.figure(figsize=(6, 6))
plt.scatter(Y, Y_pred, color='purple', label='Predictions')
plt.plot([min(Y), max(Y)], [min(Y), max(Y)], color='red', linestyle='--', label='Ideal Fit')
plt.xlabel('Actual Y')
plt.ylabel('Predicted Y')
plt.title('Predicted vs. Actual Y')
plt.legend()
plt.grid(True)
plt.show()

# C. Coefficient Paths as Lambda Varies
# Compute coefficients for each lambda
coefficients = []
for lambd in lambdas:
    beta = ridge_regression(X, Y, lambd)
    coefficients.append(beta)

coefficients = np.array(coefficients)

plt.figure(figsize=(8, 6))
plt.plot(lambdas, coefficients[:, 1], marker='o', label='Beta1 (X1)')
plt.plot(lambdas, coefficients[:, 2], marker='s', label='Beta2 (X2)')
plt.xscale('log')
plt.xlabel('Lambda (Regularization Parameter)')
plt.ylabel('Coefficient Value')
plt.title('Ridge Regression Coefficients vs. Lambda')
plt.legend()
plt.grid(True)
plt.show()

# D. Scatter Plot with Regression Line
plt.figure(figsize=(8,6))
plt.scatter(X1, Y, color='blue', label='Data Points')

# Generate a range of X1 values for plotting the regression line
X1_range = np.linspace(min(X1), max(X1), 100)
# Compute corresponding X2 assuming X2 = 3 * X1 (without noise for a cleaner line)
X2_range = 3 * X1_range
# Compute predicted Y using the final coefficients
Y_range = final_beta[0] + final_beta[1] * X1_range + final_beta[2] * X2_range

plt.plot(X1_range, Y_range, color='red', label='Ridge Regression Line')
plt.xlabel('Feature X1')
plt.ylabel('Target Y')
plt.title('Feature X1 vs. Target Y with Ridge Regression Line')
plt.legend()
plt.grid(True)
plt.show()
```
